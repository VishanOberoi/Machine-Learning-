For my understanding

Boxplot 

The box of the boxplot is the interquartile range of the data. Boxplots are good for finding out the outliers of a particular feature. The line in the middle is the median. 

Stripplot can be thought of as a complement to violin/ boxplot.

```python
ax = sns.boxplot(x ='Type' , y ='Compactness' , data= seed)
ax = sns.stripplot(x ='Type' , y ='Compactness' , jitter = True,data= seed)
```

```python
sns.violinplot(x ='Type' , y ='Compactness' , data= seed)
```

```python
#KDE PLOT

#Checking for normal distribution via shape
sns.FacetGrid(seed,hue = "Type", size = 8).map(sns.kdeplot,"Compactness").add_legend()
```

```python
#KDE PLOT

sns.FacetGrid(seed,hue = "Type", size = 8).map(sns.kdeplot,"Area").add_legend()
```

```python
#PAIR PLOT
sns.pairplot(seed,hue = "Type",diag_kind = "hist")
```
